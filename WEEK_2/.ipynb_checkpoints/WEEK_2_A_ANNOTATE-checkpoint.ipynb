# SEMAINE 2 - ETAPE 1 : L'ANNOTATION
*Marie Bizais*

## Remarques sur l'expérience d'annotation
Après une courte expérience d'annotation, force est de constater que :
- l'annotation n'a pas été homogène
- des contributeurs ou contributrices ont ajouté de nouvelles catégories
- des lignes ont été oubliées/négligées
- les zones sélectionnées ne couvraient pas toute la zone inscrite (lettres ou signes de ponctuation tronqués).

*Réfléchissons un instant sur les raisons qui motivent une annotation et sur quelques pratiques partagées en la matière.*

## Pourquoi annoter ?
L'annotation est chronophage et exige de la rigueur, mais elle s'avère souvent nécessaire. Par exemple, notre modèle de détection YOLO identifie, par défaut, un réfrigérateur correspondant au cadre de chaque image (un rectangle blanc-crème). Cela ne correspond pas à ce que nous souhaitons identifier.

### a. Etablir un modèle
On annote tout d'abord pour définir ce que nous attendons en sortie de traitement, par exemple :
- délimiter des lignes ou des colonnes
- délimiter des objets
- classifier des éléments en fonction d'une typologie que nous aurions définie
- transcrire du texte selon des normes que nous aurons choisies
- produire un tableau structuré d'informations extraites d'un texte.

*Nos annotations serviront de modèle. Soit nous entraînerons un modèle pour qu'il 'se modèle' sur la référence que nous établissons. Soit nous présenterons ces annotations à un grand modèle de langue génératif pour qu'il prenne nos exemples comme modèle à suivre.*

### b. Saisir plus précisément ses besoins
Cette démarche nous oblige à définir précisément nos besoins qui se reflèteront dans nos choix d'annotation.

Pour prendre l'exemple des pages de livres annotés :
- de quelles catégories de lignes avons-nous besoin ?
- souhaitons-nous récupérer les rares inscriptions manuscrites sur les pages ?

Dans le cas d'une transcription de texte :
- procèderons-nous à une standardisation (par exemple '&' = 'et') ?
- déplierons-nous les abbréviations pour que 'saint' = 'st' = 'St' = 'st.' ?

*Nous aurons besoin d'établir un cahier des charges.*

## Combien annoter ?
Le volume des données à annoter est très variable.

S'il s'agit de fournir quelques exemples, on tiendra compte de la variété de nos données de départ pour que les exemples couvrent autant que possible les normes que nous souhaitons voir respecter.

S'il s'agit d'ajuster un modèle dont nous savons qu'il répond déjà partiellement à nos besoins, un raffinement basé sur quelques dizaines d'exemples suffira.

Si le champ que nous abordons est nouveau, il faudra un apport massif de données (même si cela est très relatif et varie beaucoup). Il existe plusieurs stratégies dans ce cas :
- produire des données à la main
- s'appuyer sur des données préexistantes (et les adapter au besoin)
- générer des données synthétiques à partir des données dont nous disposons
- annoter à l'aide des modèles génératifs et quelques exemples, corriger les annotations ainsi produites si nécessaire.

*Notre cas d'étude correspond à la deuxième situation : YOLO propose des modèles polyvalents, nous annotons 50 pages.*

## Comment annoter ?
Il n'y a pas de règle, mais il y a quelques principes directeurs pour cette étape-clef du travail :
- On fera un point sur des projets similaires, et on tiendra compte des normes respectées dans son contexte disciplinaire.
- On réfléchira à la spécificité de nos besoins : vise-t-on à éditer un corpus ? à recenser les variantes dans diverses éditions ? à chercher des informations dans un corpus qu'on souhaiterait homogène d'un point de vue orthographique ?
- Il est recommandé de travailler à plusieurs pour que les choix soient concscients et discutés et pour que les règles soient partagées, partageables et claires.
- On commencera par annoter doublement un petit jeu de données -- ou au moins par le relire avec un deuxième oeil -- pour identifier des oublis et préciser les modalités d'annotation.
- On mettra les règles retenues par le projet par écrit sous la forme d'un cahier des charges qu'il est recommandé de présenter dans la documentation du projet (ces informations sont importantes pour que les données produites puissent être correctement utilisées).

## Retour à notre cas
Nous disposons d'un recueil de poésies. *Nous identifions un titre courant en haut des pages, celui qui correspond soit au recueil soit à la série de poèmes selon la section. Un titre courant n'est pas un élément que le lecteur lira sur la page, c'est une métadonnée qui lui sert à se repérer à l'intérieur du recueil qu'il/elle consulte.*

Nous ne voulons pas nous fermer la possibilité d'en proposer une édition numérique, page à page, ou poème par poème. *Donc nous avons besoin de récupérer les numéros de pages qui figurent en haut des pages. En revanche, nous pouvons ignorer la foliation ajoutée à la main par des employés de la bibliothèque dans le cadre de la numérisation.*

L'ouvrage n'est pas seulement composé d'une série de poèmes : il comprend également une section critique, et des corrections éditoriales. *Nous souhaitons distinguer la prose des poèmes. Nous ne savons pas si nous ferons quelque chose des corrections éditoriales. Comment les traiter ?*

**Nous devons être attentifs à une chose très importante : plus nous augmenterons le nombre de catégories, plus nous rendrons difficile l'apprentissage du modèle. Nous devons en outre être capables d'identifier visuellement à quel type de ligne correspond une catégorie, sinon, la catégorie ne sera pas 'saisissable' par le modèle YOLO.**

Nous identifions également des lieux et des dates, souvent justifiés à droite, mais relativement rares. *Peut-être faut-il associer toutes les informations avec ce type de disposition sur la page à une seule catégorie ?*

**L'identification des lignes de texte vise avant tout à extraire le texte des images. Il est donc essentiel que les cadres ou rectangles que nous traçons autour de ces lignes de texte incluent l'intégralité du texte, sans rogner aucune lettre ni aucun signe de ponctuation.** Nous pouvons déborder sur d'autres cadres ou rectangles, cela n'a pas grande importance.

## Proposition pour la typologie des lignes

On appelle ligne le rectangle (*bounding box*) qu'on trace sur la page autour d'une ligne de texte.

On distinguera les catégories suivantes :
- sur le haut de la page :
    - le numéro de page : celui qui est imprimé sur la page
    - le titre courant : titre du recueil ou de la section
- dans la partie principale de la page (la partie lue immédiatement) :
    - le titre principal : il correspond au titre du poème en début de poème ou au titre de d'une section ou d'une partie du recueil.
    - les vers : les vers des poèmes
    - la prose : les lignes en prose, qui ne sont pas alignées de la même façon que les vers (plus à gauche sur la page)
    - les données additionnelles : les lignes de texte, souvent en prose, parfois des vers ou factuelles, qui sont alignées à droite
    - les intelignes-astériques : les interlignes marqués par trois astérisques qui forment un triangle, un chiffre ou tout signe typographie utilisé pour séparer des parties d'un (long) poème.

On ne distinguera donc pas de page de titre.

Les pages de correction ne seront pas évidentes à traiter et demanderont une attention particulière. On étiquettera "prose" les lignes justifiées à gauche, "vers" les lignées indentées, "donnéesadd" les lignes justifiées à droite.

On veillera toujours à ce que le rectangle englobe la totalité du texte transcrit sur la ligne, même si cette ligne est courbe, même si un signe de ponctuation étrange figure en début ou en fin de ligne.

*PS : Ceci est une proposition. On aurait aussi pu choisir d'opérer au niveau du paragraphe, de la strophe, du titre.*
