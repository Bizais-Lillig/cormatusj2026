## Préparation de l'environnement
### Les librairies

````python
import pytesseract
from PIL import Image
from pathlib import Path
from collections import defaultdict
````

### Les fichiers

````python
PRED_DIR = Path("./predictions/my_model")
CROPS_DIR = PRED_DIR / "crops"
OCR_DIR   = PRED_DIR / "transcriptions"   # ← dossier de sortie
OCR_DIR.mkdir(exist_ok=True)
SUFFIXES = [".jpg", ".jpeg", ".JPG", ".JPEG", ".png", ".PNG"]
````

## Traitement

````python
groups = defaultdict(list)
for file in CROPS_DIR.iterdir():
    if file.suffix.lower() not in SUFFIXES:
        continue
    stem_parts = file.stem.rsplit("_Verse", 1)
    if len(stem_parts) != 2:
        continue
    base_name = stem_parts[0]           # "image_34"
    verse_num = int(stem_parts[1])      # 3
    groups[base_name].append((verse_num, file))

# -- Transcription dans l'ordre pour chaque image
for base_name, crops in sorted(groups.items()):
    crops.sort(key=lambda x: x[0])     # tri par numéro de vers

    lines = []
    for verse_num, file in crops:
        img = Image.open(file)
        text = pytesseract.image_to_string(img, lang="fra").strip()
        lines.append(text)

    out_path = OCR_DIR / f"{base_name}_Verses_transcr.txt"
    out_path.write_text("\n".join(lines), encoding="utf-8")
    print(f"✓ {out_path.name}  ({len(crops)} verse(s))")

print(f"\nTranscriptions sauvegardées dans : {OCR_DIR}/")
````
