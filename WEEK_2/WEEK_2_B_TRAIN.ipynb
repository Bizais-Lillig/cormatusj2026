## Préparation de l'enironnement
### Librairies et modèle

```python
from ultralytics import YOLO
```

```python
model = YOLO('yolo12s.pt')
```

### Fichiers

```python
train_path = "../DATA/test_yolo/data.yaml"
```

### Configuration

## Entraînement

```python
model.train(data = train_path, flipud = 0, dropout = 0.1, epochs = 50, batch = 4, imgsz = 1024, seed = 42, half = True, project='./runs', name='entrainement', device = 0)
```

## Résultats

```python
mymodel = YOLO('./entrainement/weights/best.pt')
```

```python
metrics = mymodel.val(
    data=train_path,
    split='val',
    imgsz=1024,
    batch=8,
    device='0',
    plots=False,
    save_json=False,
)
```
