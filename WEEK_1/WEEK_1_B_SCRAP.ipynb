## 2. Récupération des images

### 2.1. On établit la liste des images
On s'appuie sur le manifeste ``iiif`` :
- on récupère et examine le manifeste
- on dresse la liste des urls où sont les images
- on récupère le contenu sur chacune de ces pages
- on l'enregistre au format ``jpg`` dans le dossier prévu à cet effet

On laisse passer un peu de temps entre chaque collecte pour éviter d'être bloqué.

Pour accéder au manifeste et ne pas recevoir une réponse 403 : https://scrape.do/blog/python-requests-403-forbidden/

````python

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
    'Accept-Language': 'en-US,en;q=0.5',
    'Accept-Encoding': 'gzip, deflate',
    'Connection': 'keep-alive',
}
````

````python
#resultat = requests.get(manifest_corm)
resultat = requests.get(manifest_corm, headers=headers)
resultat #on attend une réponse 200, sinon c'est un message d'erreur
````

````python
manifest = resultat.json()
manifest#étudier le manifeste pour s'assurer que j'arrive à la liste des fichiers
````

````python
#manifest.get('sequences',[])[0].get('canvases',[])
````

````python
images_urls = []
for canvas in manifest.get('sequences',[])[0].get('canvases',[]): # on ajoute listes
    images = canvas.get('images',[])
    for image in images:
        resource = image.get('resource',[])
        service = resource.get('service')
        if service.get('@id'):
            images_urls.append(service['@id'])
print(images_urls)
````

### 3.2. On récupère à nouveau ces fichiers

````python
session = requests.Session()
session.headers.update(headers) 
durees = [10,20,30,40,50]
pause = random.choice(durees)
for pos, id in enumerate(tqdm(liste_bis)):
    try:
        r = session.get(f"https://gallica.bnf.fr/iiif/ark:/12148/bpt6k9106176d/f{id}/full/full/0/native.jpg")
        output_file = os.path.join(path_corm, f"image_{id}_bis.jpg")
        with open(output_file, "wb") as f:
            f.write(r.content)
            time.sleep(pause)
    except requests.HTTPError as e:
        print(f"Erreur http: {e}")
````

### 3.4. On renomme les fichiers mis à jour

````python
for nom in os.listdir(path_corm):
    if re.search(r'_bis\.jpg$', nom):
        nouveau_nom = nom.replace("_bis", "")
        os.rename(
            os.path.join(path_corm, nom),
            os.path.join(path_corm, nouveau_nom)
        )
        print(f"{nom} → {nouveau_nom}")
````
