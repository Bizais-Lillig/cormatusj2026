## Cadrage scientifique
Scénario :

*Nous sommes une équipe de chercheurs et nous intéressons à la littérature libanaise et à l'émergence d'un mouvement de revendication d'une identité et d'un droit à l'indépendance du liban. Notre question : comment l'espace littéraire incarne-t-il cette modernité au Liban ?*

*Nous ne souhaitons pas travailler sur des extraits ("études de cas") mais sur de larges corpus afin que nous ne partions pas de thèses a priori mais des caractéristiques que nous parviendrons à faire émerger de notre corpus.*

*Cela ne nous dispense bien sûr pas de lire sur le sujet : si le cadrage épistémologique change (nous partons des données pour attester ou infirmer des théories), la méthodologie des SHS reste fondamentale. Nous devons connaître quelles thèses et théories sont associées à notre corpus. Nous devons aussi choisir des angles d'analyse (narratologique, sémiotique, actancielle, ou autre)."*

*Ce travail pourra s'accompagner d'un travail éditorial, car nous constatons, en initiant nos travaux, que peu de textes libanais de cette période sont disponibles et utilisables par la recherche (vous pouvez chercher les auteurs les plus connus sur Wikisource pour vous en assurer).*

*Nous devons d'abord choisir un premier auteur sur lequel travailler. Notre tâche :*
- identifier la période sur laquelle nous pourrons travailler en fonction des droits d'auteurs qui s'appliquent, car même si nous ne publions pas les textes en ligne, nous devons pouvoir publier de larges extraits obtenus lors du traitement du corpus.
- identifier des auteurs éligibles pour en sélectionner un ou plusieurs.

*Compte tenu du cadre retenu et du fait que c'est notre première expérience, nous choisirons un texte francophone (un auteur arabophone posera probablement plus de difficultés en raison de la prédominance des langues européennes que sont l'anglais, le français et l'allemand dans le traitement automatique de la langue.*

*Nous nous demanderons également comment nous partagerons les résultats de notre travail :*
- s'il s'agit d'un article, quelle revue ?
- s'il s'agit d'une édition, en ligne ou papier ? Avec quels moyens ? Qui aura le droit de copier, coller, modifier ce que nous publierons ?
- s'il s'agit du code utilisé pour le traitement de nos données, le publierons-nous ? Qui pourra le réutiliser ? A quelles conditions ?
- si nous constituons des ressources complémentaires autour des textes (bibliographies, base de données des personnages, etc), les publierons-nous ? Comment ? Dans quel(s) format(s) ? Qui aura le droit de les réutiliser ? A quelles conditions ?
