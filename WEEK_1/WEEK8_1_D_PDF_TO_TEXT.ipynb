## 1. Préparation de l'environnement de travail
### 1.1. Outils: imports de librairies

````python
import pdfplumber
import os
````

### 1.2. Source et espace de sauvegarde

````python
input = "/home/bizais/Documents/Distam/2026/cormatusj2026/DATA/Volcan-Corm-extrait.pdf"
output_dir = "../DATA/texte/Corm_volcan"
os.makedirs(output, exist_ok=True)
````

## 2. Traitement des données
### 2.1. Avec vérification

````python
with pdfplumber.open(input) as pdf:
    texte_complet = ""
    for page in pdf.pages:
        texte_complet += page.extract_text()

print(texte_complet)
````

### 2.2. Sauvegarde du texte brut

````python
with open(os.path.join(output_dir,"Corm_volcan.txt"), "w", encoding="utf-8") as f:
    f.write(texte_complet)
````
